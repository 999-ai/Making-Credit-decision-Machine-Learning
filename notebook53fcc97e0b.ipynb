```python
import pandas as pd
framework=pd.read_csv("/kaggle/input/krediveriseti/krediVeriseti.csv",sep=";")
framework
```

```python
from sklearn.preprocessing import LabelEncoder
ownership=LabelEncoder()
telephone=LabelEncoder()
credit_decision=LabelEncoder()
```

```python
framework["evDurumu"]=ownership.fit_transform(framework["evDurumu"])
framework["telefonDurumu"]=telephone.fit_transform(framework["telefonDurumu"])
framework["KrediDurumu"]=credit_decision.fit_transform(framework["KrediDurumu"])
framework
```

```python

X=framework.drop("KrediDurumu",axis=1)
Y=framework["KrediDurumu"]
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(n_estimators=100,random_state=42)

model.fit(X_train,Y_train)
```

```python
Y_prediction=model.predict(X_test)
Y_prediction
```

```python
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np

mse=mean_squared_error(Y_test,Y_prediction)

print(mse,"mean squared error")
```

```python
r2_score=r2_score(Y_test,Y_prediction)
print(r2_score,"R-SQUARED")
```

```python
import matplotlib.pyplot as plt
plt.plot(Y_test,color="b",label="real values")
plt.plot(Y_prediction,color="red",label="estimation")
plt.legend()
plt.show()
```
